```python
from collections import deque

Map = {
    "Arad": [("Zerind", 75), ("Timisoara", 118), ("Sibiu", 140)],
    "Zerind": [("Oradea", 71), ("Arad", 75)],
    "Timisoara": [("Lugoj", 111), ("Arad", 118)],
    "Sibiu": [("Fagaras", 99), ("Rimmnicu Vilcea", 80), ("Arad", 140), ("Oradea", 151)],
    "Oradea": [("Zerind", 71), ("Sibiu", 151)],
    "Lugoj": [("Timisoara", 111), ("Mehadia", 70)],
    "Fagaras": [("Sibiu", 99), ("Bucharest", 211)],
    "Rimmnicu Vilcea": [("Sibiu", 80), ("Pitesti", 97), ("Craiova", 146)],
    "Pitesti": [("Rimmnicu Vilcea", 97), ("Bucharest", 101)],
    "Mehadia": [("Lugoj", 70), ("Dobreta", 75)],
    "Bucharest": [("Fagaras", 211), ("Pitesti", 101), ("Giurgiu", 90), ("Urziceni", 85)],
    "Dobreta": [("Mehadia", 75), ("Craiova", 120)],
    "Craiova": [("Dobreta", 120), ("Rimmnicu Vilcea", 146), ("Pitesti", 138)],
    "Giurgiu": [("Bucharest", 90)],
    "Urziceni": [("Bucharest", 85), ("Vasiui", 142), ("Hirsova", 98)],
    "Vasiui": [("Urziceni", 142), ("Iasi", 92)],
    "Hirsova": [("Urziceni", 98), ("Eforie", 86)],
    "Iasa": [("Vasiui", 92), ("Neamt", 87)],
    "Neamt": [("Iasa", 87)],
    "Eforie": [("Hirsova", 86)]
}

def dfs_paths(Map, start, end, path=[], weight=0):
    path = path + [(start, weight)]
    if start == end:
        return path
    shortest_path = None
    for neighbor_data in Map[start]:
        if len(neighbor_data) == 2:
            neighbor, neighbor_weight = neighbor_data
        elif len(neighbor_data) == 1:
            neighbor = neighbor_data[0]
            neighbor_weight = 0
        if neighbor not in [vertex for vertex, _ in path]:
            new_path = dfs_paths(Map, neighbor, end, path, neighbor_weight)
            if new_path:
                if not shortest_path or sum([weight for _, weight in new_path]) < sum([weight for _, weight in shortest_path]):
                    shortest_path = new_path
    return shortest_path


def bfs_paths(Map, start, end):
    queue = deque([(start, [(start, 0)])])
    shortest_path = None
    while queue:
        vertex, path = queue.popleft()
        if vertex == end:
            if not shortest_path or sum([weight for _, weight in path]) < sum([weight for _, weight in shortest_path]):
                shortest_path = path
        for neighbor_data in Map.get(vertex, []):
            if len(neighbor_data) == 2:
                neighbor, neighbor_weight = neighbor_data
            elif len(neighbor_data) == 1:
                neighbor = neighbor_data[0]
                neighbor_weight = 0
            else:
                raise ValueError(f"Invalid neighbor data for {vertex}: {neighbor_data}")
            if neighbor not in [vertex for vertex, _ in path]:
                queue.append((neighbor, path + [(neighbor, neighbor_weight)]))
    if shortest_path:
        return shortest_path
    else:
        raise ValueError(f"No path found between {start} and {end}")

def shortest_path(Map, start, end):
    dfs_shortest = dfs_paths(Map, start, end)
    bfs_shortest = bfs_paths(Map, start, end)
    if not dfs_shortest:
        return bfs_shortest
    elif not bfs_shortest:
        return dfs_shortest
    else:
        dfs_weight = sum([weight for _, weight in dfs_shortest])
        bfs_weight = sum([weight for _, weight in bfs_shortest])
        if dfs_weight < bfs_weight:
            return dfs_shortest
        else:
            return bfs_shortest
        

start = "Arad"
end = "Bucharest"
shortest = shortest_path(Map, start, end)

print("Shortest path:", shortest)

dfs_shortest = dfs_paths(Map, start, end)
bfs_shortest = bfs_paths(Map, start, end)
if not dfs_shortest:
    print("DFS could not find a path.")
elif not bfs_shortest:
    print("BFS could not find a path.")
else:
    dfs_weight = sum([weight for _, weight in dfs_shortest])
    bfs_weight = sum([weight for _, weight in bfs_shortest])
    if dfs_weight < bfs_weight:
        print("DFS found the shortest path.")
        print()
        print(f"The total weight calculated through DFS: ", dfs_weight)
    else:
        print("BFS found the shortest path.")
        print()
        print(f"The total weight calculated through BFS: ", bfs_weight)
```
